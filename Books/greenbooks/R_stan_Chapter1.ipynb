# Chapter3 確率の基本

## 標本空間と事象


試行：不確実性が伴う観測や実験

標本空間：試行によって起こりうるすべての結果の集合

事象：標本空間の部分集合

$A \cup B$：事象Aと事象Bの和集合<br>
$A \cap B$：事象Aと事象Bの積集合<br>

排反事象：事象Aと事象Bに要素の重複がないこと

$\emptyset$：要素のない事象（空事象）


## 確率

#### 確率の公理主義的定義
1. すべての事象$A$に対して$0 \leq P(A) \leq 1$ 事象が起こる確率は0以上１以下である
2. $P(\Omega) = 1$　標本空間（起こり得るすべての結果）を対象としたら、それが起こる確率は１
3. 排反事象$A_{1},A_{2},A_{3},...$に対して、$P(A_{1} \cup A_{2} \cup A_{3} \cup ...) = P(A_{1})+P(A_{2})+P(A_{3})+...$
排反事象のどれかが起こる確率は、その事象が起こる確率の和となる


## 確率の加法定理
排反事象$A_{1}$,$A_{2}$に対して、
$P(A_{1} \cup A_{2}) = P(A_{1}) + P(A_{2})$
が成り立つことを、確率の加法定理という

排反でない事象$A$,$B$において、AまたはBである確率は以下のように求められる。<br>
$P(A \cup B) = P(A) + P(B) - P(A \cap B)$

## 条件付き確率

事象Bが起きた下での事象Aの条件付き確率
$$ P(A|B) = \frac{P(A\cap B)}{P(B)}$$

$P(A|B)$は$P(A)$より大きくなることもあれば小さくなることもある


## 確率の乗法定理

事象$A$,$B$に対して、
$P(A \cap B) = P(A|B)P(B)$
が成り立つことを、確率の乗法定理という

A<u>かつ</u>Bとなる確率 = Bという条件下でAが起こる確率 $\times$ 本当にBになる確率



## 独立

事象$A$,$B$に対して、A,Bが独立であるならば、
$P(A \cap B) = P(A)P(B)$
となる($ P(A|B) = P(A)$とみなせる)
